```python
from pyautogui import *
import pyautogui
import time
import keyboard
import numpy as np
import random
#import wxwidgets
```

```python
# GLOBALS
training_menu = (129,626)
indy_training_area = (1149, 151)
first_player_area = (828, 419)
top_right_praise_button = (3488, 261)
praise_button = (1809, 1102)
end_qc_top_right = (2690, 382)
hover_duration = 1
players_to_praise = 10
y_iter = 93
```

```python
def my_move(coord_tuple):
    x, y = coord_tuple
    pyautogui.moveTo(x, y, duration=hover_duration)


def my_click():
    pyautogui.mouseDown()
    time.sleep(0.1)
    pyautogui.mouseUp()
```

```python
# To display mouse position
pyautogui.displayMousePosition()
```

```python
# To swap to app
time.sleep(5)

# Navigate to Training Menu
my_move(training_menu)
my_click()

# Click on Individual Training
time.sleep(2)
my_move(indy_training_area)
my_click()
```
