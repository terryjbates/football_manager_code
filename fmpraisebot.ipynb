```python
import requests
import random
import pathlib
import urllib.request
import os.path
from lxml import etree
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

playerNames = []
playerNos = []
playerImageLinks = []
teamNames = []
teamIds = []

def get_team_names_and_ids():

    leagueLinkAddress = 'https://www.transfermarkt.com.tr/super-lig/startseite/wettbewerb/TR1'

    leagueTreeResponse = requests.get(leagueLinkAddress, headers=headers)
    soupLeague = BeautifulSoup(leagueTreeResponse.content, 'html.parser')

    teamNamesTree = (etree.HTML(str(soupLeague)).xpath('/html/body/div[2]/main/div[2]/div[1]/div[2]/div[2]/div/table/tbody/tr'))

    for i in range(len(teamNamesTree)):

        if teamNamesTree[i].xpath('td[2]/a')[0].text != '':
            teamNames.append(teamNamesTree[i].xpath('td[2]/a/@href')[0].split('/')[1])
            teamIds.append(teamNamesTree[i].xpath('td[2]/a/@href')[0].split('/')[4])

def get_team_colors(teamColors):

    for i in range(len(teamNames)):

        temaColorUrl = 'https://www.transfermarkt.com.tr/' + teamNames[i] + '/datenfakten/verein/' + teamIds[i]

        tree = requests.get(temaColorUrl, headers=headers)
        soup = BeautifulSoup(tree.content, 'html.parser')

        arr = soup.findAll('p', class_='vereinsfarbe')

        if len(arr) == 0:
            randomColor1 = ("#" + ''.join([random.choice('ABCDEF0123456789') for i in range(6)])).lstrip('#')
            randomColor2 = ("#" + ''.join([random.choice('ABCDEF0123456789') for i in range(6)])).lstrip('#')

            teamColors[i][0] = tuple(int(randomColor1[i:i + 2], 16) for i in (0, 2, 4))
            teamColors[i][1] = tuple(int(randomColor2[i:i + 2], 16) for i in (0, 2, 4))

        else:
            h1 = arr[0].findAll('span')[0]["style"].split(':')[1].rstrip(";").lstrip('#')
            h2 = arr[0].findAll('span')[1]["style"].split(':')[1].rstrip(";").lstrip('#')

            teamColors[i][0] = tuple(int(h1[i:i + 2], 16) for i in (0, 2, 4))
            teamColors[i][1] = tuple(int(h2[i:i + 2], 16) for i in (0, 2, 4))

def get_player_datas(playerInfos):

    for i in range(len(teamNames)):

        playerNames.clear()
        playerImageLinks.clear()
        playerNos.clear()

        page = 'https://www.transfermarkt.com.tr/' + teamNames[i] + '/startseite/verein/' + teamIds[i]

        tree = requests.get(page, headers=headers)
        soup = BeautifulSoup(tree.content, 'html.parser')

        playerInfos[i][0] = teamNames[i]

        players = soup.findAll('img', class_='bilderrahmen-fixed')

        for j in range(len(players)):
            if players[j]['title'] != '':
                playerNames.append(players[j]['title'].replace(" ", "-").lower())

                playerLink = soup.find('a', attrs={'title':players[j]['title']})

                pagePlayerInfo = 'https://www.transfermarkt.com.tr' + playerLink.findNext('a')['href']
                treePlayerInfo = requests.get(pagePlayerInfo, headers=headers)
                soupPlayerInfo = BeautifulSoup(treePlayerInfo.content, 'html.parser')

                playerName = players[j]['title']

                age = soupPlayerInfo.find('span',attrs={'class':'data-header__content', 'itemprop':'birthDate'})
                randomAge = random.randint(15, 39)
                strAge = age.text.replace(" ", "").split('(')[1][:2] if age is not None else str(randomAge)
                playerAge = strAge

                number = soupPlayerInfo.find('span',attrs={'class':'data-header__shirt-number'})
                randomNumber = random.randint(0, 99)

                while randomNumber in playerNos:
                    randomNumber = random.randint(0, 99)

                playerNos.append(randomNumber)
                strNumber = number.text.replace(" ", "")[2:] if number is not None else str(randomNumber)
                playerNumber = strNumber

                value = soupPlayerInfo.find('a',attrs={'class':'data-header__market-value-wrapper'})

                thousandOrMillion = random.randint(0, 1)

                if thousandOrMillion == 1:
                    randomValue = random.randint(50, 999)
                    strValue = value.text.split('€')[0] if value is not None else str(randomValue) + " Bin "
                else:
                    randomValue = random.randint(1, 15)
                    strValue = value.text.split('€')[0] if value is not None else str(randomValue) + ".00 mil. "

                playerValue = strValue + "€"

                height = soupPlayerInfo.find('span',attrs={'class':'data-header__content', 'itemprop':'height'})
                playerHeight = height.text if height is not None else str(round(random.uniform(1.6, 2.0), 2)) + " m"
                strHeight = playerHeight.replace(" ", "").replace("m", "").replace(",", "").replace(".", "")

                playerWeight = str(round(calculate_age_constant(int(playerAge)) * (int(strHeight) - 105))) + " kg"

                playerInfos[i][1].append(playerName)
                playerInfos[i][2].append(playerAge)
                playerInfos[i][3].append(playerNumber)
                playerInfos[i][4].append(playerValue)
                playerInfos[i][5].append(playerHeight)
                playerInfos[i][6].append(playerWeight)

        for playerImageLink in soup.findAll('img', class_='bilderrahmen-fixed'):

            playerImageLinks.append(playerImageLink['data-src'].replace("small", "header"))

        for j in range(len(playerNames)):

            absolute_path = os.path.dirname(__file__)
            relative_path = "output-photos\\player-photos\\" + teamNames[i] + "\\"
            full_path = os.path.join(absolute_path, relative_path)

            if not os.path.isdir(full_path):
                pathlib.Path(full_path).mkdir(parents=True, exist_ok=True)

            urllib.request.urlretrieve(playerImageLinks[j], full_path + playerNames[j] + '.png')

def calculate_age_constant(ageValue):
    if ageValue < 15:
        return 0.9
    elif ageValue >= 15 and ageValue < 20:
        return 0.95
    elif ageValue >= 20 and ageValue < 25:
        return 1.00
    elif ageValue >= 25 and ageValue < 30:
        return 1.00
    elif ageValue >= 30 and ageValue < 35:
        return 1.00
    elif ageValue >= 35 and ageValue < 40:
        return 1.05
    else:
        return 1.1

def create_team_card_templates(teamColors):

    for i in range(len(teamNames)):

        absolute_path = os.path.dirname(__file__)

        flag_inside1 = Image.open(absolute_path + "\\base-photos\\flag-inside-1.png")
        flag_inside2 = Image.open(absolute_path + "\\base-photos\\flag-inside-2.png")
        inline_border = Image.open(absolute_path + "\\base-photos\\inline-border.png")

        new_image1 = []
        new_image2 = []
        new_image3 = []

        flag_inside1 = flag_inside1.convert("RGBA")
        flag_inside2 = flag_inside2.convert("RGBA")
        inline_border = inline_border.convert("RGBA")

        d1 = flag_inside1.getdata()
        d2 = flag_inside2.getdata()
        d3 = inline_border.getdata()

        color1 = teamColors[i][0]
        color2 = teamColors[i][1]

        for item in d1:

            if item[0] in list(range(200, 256)):
                new_image1.append(color1)
            else:
                new_image1.append(item)

        for item in d2:

            if item[0] in list(range(200, 256)):
                new_image2.append(color2)
            else:
                new_image2.append(item)

        for item in d3:

            if item[0] in list(range(200, 256)):
                new_image3.append(color1)
            else:
                new_image3.append(item)

        flag_inside1.putdata(new_image1)
        flag_inside2.putdata(new_image2)
        inline_border.putdata(new_image3)

        relative_path = "output-photos\\team-templates\\" + teamNames[i] + "\\"
        full_path = os.path.join(absolute_path, relative_path)

        if not os.path.isdir(full_path):
            pathlib.Path(full_path).mkdir(parents=True, exist_ok=True)

        flag_inside1.save(full_path + "flag-inside1.png")

        flag_inside2.save(full_path + "flag-inside2.png")

        inline_border.save(full_path + "inline-border.png")

def generate_team_cards(playerInfos):

    for i in range(len(playerInfos)):

        for j in range(len(playerInfos[i][1])):

            absolute_path = os.path.dirname(__file__)

            background = Image.open(absolute_path + "\\base-photos\\background.png")
            base_info = Image.open(absolute_path + "\\base-photos\\base-info.png")
            flag_border = Image.open(absolute_path + "\\base-photos\\flag-border.png")

            flag_inside1 = Image.open(absolute_path + "\\output-photos\\team-templates\\" + playerInfos[i][0] + "\\" + "flag-inside1.png")
            flag_inside2 = Image.open(absolute_path + "\\output-photos\\team-templates\\" + playerInfos[i][0] + "\\" + "flag-inside2.png")
            inline_border = Image.open(absolute_path + "\\output-photos\\team-templates\\" + playerInfos[i][0] + "\\" + "inline-border.png")

            photo = Image.open(absolute_path + "\\output-photos\\player-photos\\" + playerInfos[i][0] + "\\" + playerInfos[i][1][j].replace(" ", "-").lower() + ".png")

            photo = photo.resize((550, 700))

            im = Image.new(mode="RGBA", size=(background.width, background.height))

            im.paste(photo, (150, 110))

            relative_path = "output-photos\\resized-player-photos\\" + playerInfos[i][0] + "\\"

            full_path = os.path.join(absolute_path, relative_path)

            if not os.path.isdir(full_path):
                pathlib.Path(full_path).mkdir(parents=True, exist_ok=True)

            im.save(full_path + playerInfos[i][1][j].replace(" ", "-").lower() + ".png")

            photo = Image.open(full_path + playerInfos[i][1][j].replace(" ", "-").lower() + ".png")

            img1 = Image.alpha_composite(background, photo)
            img2 = Image.alpha_composite(img1, base_info)
            img3 = Image.alpha_composite(img2, inline_border)
            img4 = Image.alpha_composite(img3, flag_inside1)
            img5 = Image.alpha_composite(img4, flag_inside2)
            img6 = Image.alpha_composite(img5, flag_border)

            i1 = ImageDraw.Draw(img6)

            nameFont = ImageFont.truetype(absolute_path + '\\fonts\\arial-narrow.ttf', 60)
            infoFont = ImageFont.truetype(absolute_path + '\\fonts\\UniversalisADFStd-CondItalic.otf', 53)
            numberFont = ImageFont.truetype(absolute_path + '\\fonts\\arial.ttf', 60)

            name = playerInfos[i][1][j]

            age = playerInfos[i][2][j]
            weight = playerInfos[i][6][j]
            height = playerInfos[i][5][j]
            value = playerInfos[i][4][j]
            number = playerInfos[i][3][j]

            i1.text((img6.width/2, 902), name, font=nameFont, fill=(0, 0, 0), anchor="mm")

            i1.text((img6.width/4, 1028), age, font=infoFont, fill=(0, 0, 0), anchor="mm")
            i1.text((img6.width*(3/4), 1028), weight, font=infoFont, fill=(0, 0, 0), anchor="mm")
            i1.text((img6.width/4 + 40, 1120), height, font=infoFont, fill=(0, 0, 0), anchor="mm")
            i1.text((img6.width*(3/4) + 50, 1120), value, font=infoFont, fill=(0, 0, 0), anchor="mm")
            i1.text((img6.width*(3/4) + 92, 125), number, font=numberFont, fill=(0, 0, 0), anchor="mm")

            relative_path = "output-photos\\final-results\\" + playerInfos[i][0] + "\\"

            full_path = os.path.join(absolute_path, relative_path)

            if not os.path.isdir(full_path):
                pathlib.Path(full_path).mkdir(parents=True, exist_ok=True)

            img6.save(full_path + playerInfos[i][1][j].replace(" ", "-").lower() + ".png")

def main():
    get_team_names_and_ids()

    teamColors = [[0 for i in range(2)] for j in range(len(teamNames))]
    playerInfos = [[str(), list(), list(), list(), list(), list(), list()] for j in range(len(teamNames))]

    get_team_colors(teamColors)
    get_player_datas(playerInfos)
    create_team_card_templates(teamColors)
    generate_team_cards(playerInfos)

if __name__ == "__main__":
    main()
```

```python
from __future__ import annotations

import time
from dataclasses import dataclass
from typing import Callable, Iterable, List, Optional, Sequence, Tuple

import cv2
import numpy as np
import pyautogui
import pydirectinput as pdi

Point = Tuple[int, int]
Region = Tuple[int, int, int, int]  # (left, top, width, height)


# ----------------------------
# Core CV + input primitives
# ----------------------------

def screenshot_bgr(region: Region) -> np.ndarray:
    """Screenshot a region (screen coords) and return BGR ndarray for OpenCV."""
    img = pyautogui.screenshot(region=region)
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)


def load_template(path: str) -> np.ndarray:
    tpl = cv2.imread(path, cv2.IMREAD_COLOR)
    if tpl is None:
        raise ValueError(f"Template image not found or unreadable: {path}")
    return tpl


def match_one(
    region: Region,
    template_path: str,
    threshold: float = 0.80,
    method: int = cv2.TM_CCOEFF_NORMED,
) -> Optional[Tuple[Point, float, Tuple[int, int]]]:
    """
    Return (center_screen_xy, score, (tpl_w, tpl_h)) for best match, else None.
    """
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)

    sh_h, sh_w = shot.shape[:2]
    tp_h, tp_w = tpl.shape[:2]
    if tp_h > sh_h or tp_w > sh_w:
        raise ValueError(
            f"Template ({tp_w}x{tp_h}) larger than screenshot region ({sh_w}x{sh_h})."
        )

    result = cv2.matchTemplate(shot, tpl, method)
    _, max_val, _, max_loc = cv2.minMaxLoc(result)

    if max_val < threshold:
        return None

    left, top, _, _ = region
    x0, y0 = max_loc  # top-left (x,y) within region screenshot
    center_x = left + x0 + tp_w // 2
    center_y = top + y0 + tp_h // 2
    return (center_x, center_y), float(max_val), (tp_w, tp_h)


def find_all_grouped_centers(
    region: Region,
    template_path: str,
    threshold: float = 0.70,
    group_px: int = 10,
    method: int = cv2.TM_CCOEFF_NORMED,
) -> List[Point]:
    """
    Find all matches above threshold in the region, group nearby detections,
    and return grouped centers in SCREEN coords.
    """
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)
    tp_h, tp_w = tpl.shape[:2]

    result = cv2.matchTemplate(shot, tpl, method)
    ys, xs = np.where(result >= threshold)
    if len(xs) == 0:
        return []

    # Convert raw detections (top-left in region image) -> centers in SCREEN coords
    left, top, _, _ = region
    centers = [(left + int(x) + tp_w // 2, top + int(y) + tp_h // 2) for x, y in zip(xs, ys)]

    return group_points(centers, threshold_px=group_px)


def group_points(points: Sequence[Point], threshold_px: int = 10) -> List[Point]:
    """Cluster points that are within threshold_px in both x and y; return cluster means."""
    grouped: List[List[Point]] = []
    for pt in points:
        for g in grouped:
            gx, gy = g[0]
            if abs(gx - pt[0]) < threshold_px and abs(gy - pt[1]) < threshold_px:
                g.append(pt)
                break
        else:
            grouped.append([pt])

    out: List[Point] = []
    for g in grouped:
        arr = np.array(g, dtype=np.int32)
        mean = arr.mean(axis=0).astype(int)
        out.append((int(mean[0]), int(mean[1])))
    return out


def move_and_click(pt: Point, move_duration: float = 0.15, click_delay: float = 0.02) -> None:
    pdi.moveTo(pt[0], pt[1], duration=move_duration)
    time.sleep(click_delay)
    pdi.click()


def wait_until(
    predicate: Callable[[], bool],
    timeout_s: float = 2.0,
    poll_s: float = 0.10,
) -> bool:
    """Poll predicate until it returns True or timeout."""
    deadline = time.time() + timeout_s
    while time.time() < deadline:
        if predicate():
            return True
        time.sleep(poll_s)
    return False


def wait_and_click_template(
    region: Region,
    template_path: str,
    threshold: float = 0.80,
    timeout_s: float = 3.0,
    poll_s: float = 0.15,
    move_duration: float = 0.15,
) -> Optional[Point]:
    """
    Keep trying to find a template within region until timeout; click it when found.
    Returns click point or None.
    """
    def _try_find() -> Optional[Point]:
        hit = match_one(region, template_path, threshold=threshold)
        return hit[0] if hit else None

    deadline = time.time() + timeout_s
    while time.time() < deadline:
        pt = _try_find()
        if pt:
            move_and_click(pt, move_duration=move_duration)
            return pt
        time.sleep(poll_s)
    return None


# ----------------------------
# FM bot config + workflow
# ----------------------------

@dataclass(frozen=True)
class FMUI:
    # Regions
    rating_region: Region                 # left list where green bars live
    development_region: Region            # right pane where "praise" lives
    dialog_region: Region                # pop-up dialog area (responses/buttons)

    # Templates
    green_rating_tpl: str                # e.g. 'fm_26_green_space_near_player.png'
    praise_action_tpl: str               # e.g. './praise_text_image.png'

    # Optional: additional templates (fill in later)
    # praise_confirm_tpl: str
    # dialog_option_1_tpl: str
    # dialog_option_2_tpl: str
    # dialog_close_tpl: str


class FMPraiseBot:
    def __init__(self, ui: FMUI) -> None:
        self.ui = ui
        pyautogui.PAUSE = 0.0
        pdi.PAUSE = 0.03

    # --- Navigation / page management ---

    def open_training_bookmark(self) -> None:
        # shift+9 bookmark (pydirectinput already proven to work for you)
        pdi.keyDown("shift")
        pdi.press("9")
        pdi.keyUp("shift")

    def click_individual_training_tab(self, tab_point: Point) -> None:
        move_and_click(tab_point, move_duration=0.10)

    def scroll_player_list(self, amount: int, hover_point: Point) -> None:
        pdi.moveTo(hover_point[0], hover_point[1], duration=0.10)
        pyautogui.scroll(amount)

    # --- Vision-driven actions ---

    def find_green_players(self, threshold: float = 0.70, group_px: int = 12) -> List[Point]:
        return find_all_grouped_centers(
            self.ui.rating_region,
            self.ui.green_rating_tpl,
            threshold=threshold,
            group_px=group_px,
        )

    def select_player(self, player_point: Point) -> None:
        move_and_click(player_point, move_duration=0.10)

    def click_praise_in_development(self) -> bool:
        pt = wait_and_click_template(
            self.ui.development_region,
            self.ui.praise_action_tpl,
            threshold=0.80,
            timeout_s=2.5,
            poll_s=0.12,
            move_duration=0.12,
        )
        return pt is not None

    # --- Dialog flow (stubbed for now) ---

    def handle_praise_dialog(self) -> None:
        """
        Placeholder: after clicking praise, a pop-up appears.
        Implement by chaining wait_and_click_template() calls in dialog_region.
        """
        # Example skeleton (you will replace templates/regions):
        # wait_and_click_template(self.ui.dialog_region, self.ui.dialog_option_1_tpl, threshold=0.80, timeout_s=2.0)
        # wait_and_click_template(self.ui.dialog_region, self.ui.dialog_option_2_tpl, threshold=0.80, timeout_s=2.0)
        # wait_and_click_template(self.ui.dialog_region, self.ui.dialog_close_tpl,   threshold=0.80, timeout_s=2.0)
        time.sleep(0.4)

    # --- End-to-end per-player action ---

    def praise_player_if_possible(self, player_point: Point) -> bool:
        self.select_player(player_point)
        time.sleep(0.25)  # small UI settle

        if not self.click_praise_in_development():
            return False

        self.handle_praise_dialog()
        return True

    # --- Outer loop over list pages ---

    def run_praise_pass(
        self,
        pages: int,
        scroll_each_page: int,
        list_hover_point: Point,
        green_threshold: float = 0.70,
    ) -> None:
        for page_idx in range(pages):
            players = self.find_green_players(threshold=green_threshold)
            for pt in players:
                self.praise_player_if_possible(pt)
                time.sleep(0.20)

            if page_idx < pages - 1:
                self.scroll_player_list(scroll_each_page, hover_point=list_hover_point)
                time.sleep(0.50)


# ----------------------------
# Example usage (you fill in coords/templates)
# ----------------------------

def main() -> None:
    ui = FMUI(
        rating_region=(956, 475, 135, 1587),
        development_region=(3144, 226, 645, 1900),   # your PRAISE_TEXT_REGION for now
        dialog_region=(0, 0, 0, 0),                  # TODO: set when you template the dialog

        green_rating_tpl="fm_26_green_space_near_player.png",
        praise_action_tpl="./praise_text_image.png",
    )

    bot = FMPraiseBot(ui)

    time.sleep(7.0)  # time to focus FM
    bot.open_training_bookmark()

    # If you still need to click into Individual Training tab:
    # bot.click_individual_training_tab((1094, 1356))

    # Run across multiple "pages" by scrolling the player list
    bot.run_praise_pass(
        pages=4,
        scroll_each_page=-2000,
        list_hover_point=(858, 812),
        green_threshold=0.70,
    )


if __name__ == "__main__":
    main()
```

```python
from __future__ import annotations

import time
from dataclasses import dataclass
from typing import Callable, List, Optional, Sequence, Tuple

import cv2
import numpy as np
import pyautogui
import pydirectinput as pdi

Point = Tuple[int, int]
Region = Tuple[int, int, int, int]  # (left, top, width, height)


# ----------------------------
# Core CV + input primitives
# ----------------------------

def screenshot_bgr(region: Region) -> np.ndarray:
    img = pyautogui.screenshot(region=region)
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)


def load_template(path: str) -> np.ndarray:
    tpl = cv2.imread(path, cv2.IMREAD_COLOR)
    if tpl is None:
        raise ValueError(f"Template image not found or unreadable: {path}")
    return tpl


def match_one(
    region: Region,
    template_path: str,
    threshold: float = 0.80,
    method: int = cv2.TM_CCOEFF_NORMED,
) -> Optional[Tuple[Point, float, Tuple[int, int]]]:
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)

    sh_h, sh_w = shot.shape[:2]
    tp_h, tp_w = tpl.shape[:2]
    if tp_h > sh_h or tp_w > sh_w:
        raise ValueError(
            f"Template ({tp_w}x{tp_h}) larger than screenshot region ({sh_w}x{sh_h})."
        )

    result = cv2.matchTemplate(shot, tpl, method)
    _, max_val, _, max_loc = cv2.minMaxLoc(result)

    if max_val < threshold:
        return None

    left, top, _, _ = region
    x0, y0 = max_loc
    center_x = left + x0 + tp_w // 2
    center_y = top + y0 + tp_h // 2
    return (center_x, center_y), float(max_val), (tp_w, tp_h)


def find_all_grouped_centers(
    region: Region,
    template_path: str,
    threshold: float = 0.70,
    group_px: int = 10,
    method: int = cv2.TM_CCOEFF_NORMED,
) -> List[Point]:
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)
    tp_h, tp_w = tpl.shape[:2]

    result = cv2.matchTemplate(shot, tpl, method)
    ys, xs = np.where(result >= threshold)
    if len(xs) == 0:
        return []

    left, top, _, _ = region
    centers = [(left + int(x) + tp_w // 2, top + int(y) + tp_h // 2) for x, y in zip(xs, ys)]
    return group_points(centers, threshold_px=group_px)


def group_points(points: Sequence[Point], threshold_px: int = 10) -> List[Point]:
    grouped: List[List[Point]] = []
    for pt in points:
        for g in grouped:
            gx, gy = g[0]
            if abs(gx - pt[0]) < threshold_px and abs(gy - pt[1]) < threshold_px:
                g.append(pt)
                break
        else:
            grouped.append([pt])

    out: List[Point] = []
    for g in grouped:
        arr = np.array(g, dtype=np.int32)
        mean = arr.mean(axis=0).astype(int)
        out.append((int(mean[0]), int(mean[1])))
    return out


def move_and_click(pt: Point, move_duration: float = 0.15, click_delay: float = 0.02) -> None:
    pdi.moveTo(pt[0], pt[1], duration=move_duration)
    time.sleep(click_delay)
    pdi.click()


def wait_and_click_template(
    region: Region,
    template_path: str,
    threshold: float = 0.80,
    timeout_s: float = 3.0,
    poll_s: float = 0.15,
    move_duration: float = 0.15,
) -> Optional[Point]:
    deadline = time.time() + timeout_s
    while time.time() < deadline:
        hit = match_one(region, template_path, threshold=threshold)
        if hit:
            pt = hit[0]
            move_and_click(pt, move_duration=move_duration)
            return pt
        time.sleep(poll_s)
    return None


# ----------------------------
# FM bot config + workflow
# ----------------------------

@dataclass(frozen=True)
class FMUI:
    # Safety: what window must be active for clicks to be allowed
    window_title_contains: str

    # Regions
    rating_region: Region
    development_region: Region
    dialog_region: Region

    # Templates
    green_rating_tpl: str
    praise_action_tpl: str
    
    # NEW: required click location after Shift+9 to open the list
    indy_training_tab_point: Point


def _active_window_title() -> str:
    """Windows-only best effort; returns '' if pywin32 not installed."""
    try:
        import win32gui  # pip install pywin32
        hwnd = win32gui.GetForegroundWindow()
        return win32gui.GetWindowText(hwnd) or ""
    except Exception:
        return ""


class FMPraiseBot:
    def __init__(self, ui: FMUI) -> None:
        self.ui = ui
        pyautogui.PAUSE = 0.0
        pdi.PAUSE = 0.03

    # --- Safety / focus ---

    def focus_fm_or_die(self, settle_s: float = 0.8) -> None:
        """
        Bring FM to the foreground and verify it is the active window.
        If verification fails, raise RuntimeError so we do NOT click your notebook.
        """
        wins = pyautogui.getWindowsWithTitle(self.ui.window_title_contains)
        if not wins:
            raise RuntimeError(
                f"FM window not found: no window title contains {self.ui.window_title_contains!r}"
            )

        w = wins[0]
        try:
            if getattr(w, "isMinimized", False):
                w.restore()
            w.activate()
        except Exception:
            # We'll still verify foreground title after settle
            pass

        time.sleep(settle_s)

        active = _active_window_title()
        if self.ui.window_title_contains.lower() not in active.lower():
            raise RuntimeError(
                f"Refusing to continue: active window is {active!r} "
                f"(expected to contain {self.ui.window_title_contains!r})."
            )

    # --- Navigation / page management ---

    def open_training_bookmark(self) -> None:
        # Ensure FM is focused BEFORE sending keys
        self.focus_fm_or_die()
        pdi.keyDown("shift")
        pdi.press("9")
        pdi.keyUp("shift")

    def click_individual_training_tab(self, tab_point: Point) -> None:
        self.focus_fm_or_die()
        move_and_click(tab_point, move_duration=0.10)

    def scroll_player_list(self, amount: int, hover_point: Point) -> None:
        self.focus_fm_or_die()
        pdi.moveTo(hover_point[0], hover_point[1], duration=0.10)
        pyautogui.scroll(amount)

    # --- Vision-driven actions ---

    def find_green_players(self, threshold: float = 0.70, group_px: int = 12) -> List[Point]:
        # No click here, but still safe to keep FM focused (prevents stale state surprises)
        self.focus_fm_or_die()
        return find_all_grouped_centers(
            self.ui.rating_region,
            self.ui.green_rating_tpl,
            threshold=threshold,
            group_px=group_px,
        )

    def select_player(self, player_point: Point) -> None:
        self.focus_fm_or_die()
        move_and_click(player_point, move_duration=0.10)

    def click_praise_in_development(self) -> bool:
        self.focus_fm_or_die()
        pt = wait_and_click_template(
            self.ui.development_region,
            self.ui.praise_action_tpl,
            threshold=0.80,
            timeout_s=2.5,
            poll_s=0.12,
            move_duration=0.12,
        )
        return pt is not None

    # --- Dialog flow (stubbed for now) ---

    def handle_praise_dialog(self) -> None:
        """
        After clicking praise, a pop-up appears.
        Implement by chaining wait_and_click_template() calls in dialog_region.

        Keep this as a sequence of small, reliable "wait-and-click" steps:
          1) click the correct response option
          2) click confirm / handshake / continue (if present)
          3) click close / end chat
        """
        self.focus_fm_or_die()
        time.sleep(0.4)

    # --- End-to-end per-player action ---

    def praise_player_if_possible(self, player_point: Point) -> bool:
        self.select_player(player_point)
        time.sleep(0.25)

        if not self.click_praise_in_development():
            return False

        self.handle_praise_dialog()
        return True

    # --- Outer loop over list pages ---

    def run_praise_pass(
        self,
        pages: int,
        scroll_each_page: int,
        list_hover_point: Point,
        green_threshold: float = 0.70,
    ) -> None:
        for page_idx in range(pages):
            players = self.find_green_players(threshold=green_threshold)
            for pt in players:
                self.praise_player_if_possible(pt)
                time.sleep(0.20)

            if page_idx < pages - 1:
                self.scroll_player_list(scroll_each_page, hover_point=list_hover_point)
                time.sleep(0.50)

                
    def go_to_individual_training(self) -> None:
        """
        Required navigation step:

        1) Focus FM (safety)
        2) Hit the Shift+9 bookmark
        3) Click the Individual Training list location to actually load/activate it
        """
        self.focus_fm_or_die()

        # Bookmark
        pdi.keyDown("shift")
        pdi.press("9")
        pdi.keyUp("shift")

        time.sleep(0.8)  # give FM time to switch screen

        # REQUIRED click to open/activate the list
        move_and_click(self.ui.indy_training_tab_point, move_duration=0.2)

        time.sleep(0.8)  # let the list populate/settle
                

def main() -> None:
    ui = FMUI(
        window_title_contains="Football Manager",   # change if your title is different
        rating_region=(956, 475, 135, 1587),
        development_region=(3144, 226, 645, 1900),
        dialog_region=(0, 0, 0, 0),                 # fill in when you template the dialog
        green_rating_tpl="fm_26_green_space_near_player.png",
        praise_action_tpl="./praise_text_image.png",
        # THIS is the click you already had working:
        indy_training_tab_point=(1094, 1356),
    )


    bot = FMPraiseBot(ui)

    bot.focus_fm_or_die()
    bot.go_to_individual_training()

    bot.run_praise_pass(
        pages=4,
        scroll_each_page=-2000,
        list_hover_point=(858, 812),
        green_threshold=0.70,
    )


if __name__ == "__main__":
    main()
```

```python
from __future__ import annotations

import time
from dataclasses import dataclass
from typing import Callable, List, Optional, Sequence, Tuple

import cv2
import numpy as np
import pyautogui
import pydirectinput as pdi

Point = Tuple[int, int]
Region = Tuple[int, int, int, int]  # (left, top, width, height)


# ----------------------------
# Window focus / safety gate
# ----------------------------

def _active_window_title() -> str:
    """
    Best-effort: return active window title on Windows.
    Falls back to empty string if unavailable.
    """
    try:
        import win32gui  # pywin32
        hwnd = win32gui.GetForegroundWindow()
        return win32gui.GetWindowText(hwnd) or ""
    except Exception:
        return ""


def activate_window_contains(title_substring: str, settle_s: float = 0.8) -> None:
    """
    Bring a window containing title_substring to the foreground, then verify it is active.
    Raises RuntimeError if it cannot.
    """
    # Try via pyautogui window objects
    wins = pyautogui.getWindowsWithTitle(title_substring)
    if wins:
        w = wins[0]
        try:
            if w.isMinimized:
                w.restore()
            w.activate()
        except Exception:
            # Some windows throw; we'll still attempt fallback verify below
            pass
    else:
        raise RuntimeError(f"Could not find any window with title containing: {title_substring!r}")

    time.sleep(settle_s)

    # Verify foreground title (more reliable than assuming activate worked)
    active = _active_window_title()
    if title_substring.lower() not in active.lower():
        raise RuntimeError(
            f"Refusing to continue: active window is {active!r}, expected it to contain {title_substring!r}."
        )


def countdown_to_focus(seconds: int = 3) -> None:
    """
    Optional: gives you time to stop the bot if something looks wrong.
    """
    for i in range(seconds, 0, -1):
        print(f"Starting in {i}… (focus Football Manager now)")
        time.sleep(1)


# ----------------------------
# Core CV + input primitives
# ----------------------------

def screenshot_bgr(region: Region) -> np.ndarray:
    img = pyautogui.screenshot(region=region)
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)


def load_template(path: str) -> np.ndarray:
    tpl = cv2.imread(path, cv2.IMREAD_COLOR)
    if tpl is None:
        raise ValueError(f"Template image not found or unreadable: {path}")
    return tpl


def match_one(region: Region, template_path: str, threshold: float = 0.80) -> Optional[Tuple[Point, float]]:
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)

    sh_h, sh_w = shot.shape[:2]
    tp_h, tp_w = tpl.shape[:2]
    if tp_h > sh_h or tp_w > sh_w:
        raise ValueError(f"Template ({tp_w}x{tp_h}) larger than region ({sh_w}x{sh_h}).")

    result = cv2.matchTemplate(shot, tpl, cv2.TM_CCOEFF_NORMED)
    _, max_val, _, max_loc = cv2.minMaxLoc(result)
    if max_val < threshold:
        return None

    left, top, _, _ = region
    x0, y0 = max_loc
    center = (left + x0 + tp_w // 2, top + y0 + tp_h // 2)
    return center, float(max_val)


def group_points(points: Sequence[Point], threshold_px: int = 10) -> List[Point]:
    grouped: List[List[Point]] = []
    for pt in points:
        for g in grouped:
            gx, gy = g[0]
            if abs(gx - pt[0]) < threshold_px and abs(gy - pt[1]) < threshold_px:
                g.append(pt)
                break
        else:
            grouped.append([pt])

    out: List[Point] = []
    for g in grouped:
        arr = np.array(g, dtype=np.int32)
        mean = arr.mean(axis=0).astype(int)
        out.append((int(mean[0]), int(mean[1])))
    return out


def find_all_grouped_centers(
    region: Region,
    template_path: str,
    threshold: float = 0.70,
    group_px: int = 12,
) -> List[Point]:
    shot = screenshot_bgr(region)
    tpl = load_template(template_path)
    tp_h, tp_w = tpl.shape[:2]

    result = cv2.matchTemplate(shot, tpl, cv2.TM_CCOEFF_NORMED)
    ys, xs = np.where(result >= threshold)
    if len(xs) == 0:
        return []

    left, top, _, _ = region
    centers = [(left + int(x) + tp_w // 2, top + int(y) + tp_h // 2) for x, y in zip(xs, ys)]
    return group_points(centers, threshold_px=group_px)


def move_and_click(pt: Point, move_duration: float = 0.12) -> None:
    pdi.moveTo(pt[0], pt[1], duration=move_duration)
    pdi.click()


def wait_and_click_template(
    region: Region,
    template_path: str,
    threshold: float = 0.80,
    timeout_s: float = 3.0,
    poll_s: float = 0.15,
    move_duration: float = 0.12,
) -> Optional[Point]:
    deadline = time.time() + timeout_s
    while time.time() < deadline:
        hit = match_one(region, template_path, threshold=threshold)
        if hit:
            pt, score = hit
            print(f"Matched {template_path} score={score:.3f} at {pt}")
            move_and_click(pt, move_duration=move_duration)
            return pt
        time.sleep(poll_s)
    print(f"Timed out waiting for {template_path}")
    return None


# ----------------------------
# FM bot config + workflow
# ----------------------------

@dataclass(frozen=True)
class FMUI:
    # window
    window_title_contains: str

    # regions
    rating_region: Region
    development_region: Region
    convo_region: Region

    # templates
    green_rating_tpl: str
    praise_action_tpl: str
    praise_convo_tpl: str
    end_praise_chat_tpl: str
    
    
    # coordinates
    indy_training_tab_point: Optional[Point] = None
    list_hover_point: Optional[Point] = None


class FMPraiseBot:
    def __init__(self, ui: FMUI) -> None:
        self.ui = ui
        pyautogui.PAUSE = 0.0
        pdi.PAUSE = 0.03

    def focus_fm_or_die(self) -> None:
        activate_window_contains(self.ui.window_title_contains, settle_s=0.9)

    def open_training_bookmark(self) -> None:
        # shift+9 bookmark
        pdi.keyDown("shift")
        pdi.press("9")
        pdi.keyUp("shift")

    def click_individual_training_tab(self) -> None:
        if self.ui.indy_training_tab_point:
            move_and_click(self.ui.indy_training_tab_point, move_duration=0.10)

    def scroll_player_list(self, amount: int) -> None:
        if not self.ui.list_hover_point:
            raise RuntimeError("list_hover_point is not set; cannot safely scroll.")
        pdi.moveTo(self.ui.list_hover_point[0], self.ui.list_hover_point[1], duration=0.10)
        pyautogui.scroll(amount)

    def find_green_players(self) -> List[Point]:
        return find_all_grouped_centers(
            self.ui.rating_region,
            self.ui.green_rating_tpl,
            threshold=0.70,
            group_px=12,
        )

    def praise_player(self, player_point: Point) -> bool:
        # SAFETY: re-assert FM focus before each click burst
        self.focus_fm_or_die()

        move_and_click(player_point, move_duration=0.10)
        time.sleep(0.25)

        clicked = wait_and_click_template(
            self.ui.development_region,
            self.ui.praise_action_tpl,
            threshold=0.80,
            timeout_s=2.5,
        )
        if not clicked:
            return False

        
        
        clicked_convo_choice = wait_and_click_template(
            self.ui.convo_region,
            self.ui.praise_convo_tpl,
            threshold=0.80,
            timeout_s=2.5,
        )
        if not clicked_convo_choice:
            return False
        
        clicked_end_convo = wait_and_click_template(
            self.ui.convo_region,
            self.ui.end_praise_chat_tpl,
            threshold=0.80,
            timeout_s=2.5,
        )
        if not clicked_end_convo:
            return False
        

        # TODO: dialog handling (later)
        time.sleep(0.25)
        return True

    def run_praise_pass(self, pages: int, scroll_each_page: int) -> None:
        for page in range(pages):
            self.focus_fm_or_die()

            players = self.find_green_players()
            print(f"Page {page+1}/{pages}: found {len(players)} green candidates")

            for pt in players:
                self.praise_player(pt)
                time.sleep(0.15)

            if page < pages - 1:
                self.scroll_player_list(scroll_each_page)
                time.sleep(0.6)


def main() -> None:
    ui = FMUI(
        window_title_contains="Football Manager",  # adjust if needed: "Football Manager 2026"
        rating_region=(956, 475, 135, 1587),
        development_region=(3144, 226, 645, 1900),
        convo_region=(2355, 408, 637, 1398),
        green_rating_tpl="fm_26_green_space_near_player.png",
        praise_action_tpl="./praise_text_image.png",
        praise_convo_tpl = "./praise_convo_choice.png",
        end_praise_chat_tpl = "./end_praise_chat.png",
        indy_training_tab_point=(1094, 1356),  # optional
        list_hover_point=(858, 812),
    )

    bot = FMPraiseBot(ui)

    countdown_to_focus(5)
    bot.focus_fm_or_die()

    bot.open_training_bookmark()
    time.sleep(0.8)
    bot.click_individual_training_tab()
    time.sleep(0.8)

    # Example: 4 pages, scroll down each time
    bot.run_praise_pass(pages=4, scroll_each_page=-2000)


if __name__ == "__main__":
    main()
```
